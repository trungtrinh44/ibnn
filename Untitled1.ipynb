```
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import ticker, cm
%matplotlib inline
```

```
from google.colab import drive
drive.mount('/content/drive')
```

```
result = np.load('/content/drive/My Drive/final_submit/vgg_loss_surface/weight_space_1_2.npz')
```

```
train_loss = result['train_loss']
train_err = result['train_err']
test_loss = result['test_loss']
test_err = result['test_err']
trajectory_projection = result['trajectory_projection']
```

```
dist = 30
N = 31
xs = np.linspace(-dist, dist, N)
ys = np.linspace(-dist, dist, N)
```

```
fig, axes = plt.subplots(figsize=(10,8), ncols=2, nrows=2)
ax = axes[0][1]
cs = ax.contourf(xs, ys, train_loss, cmap=cm.Spectral_r, 
                 levels = 20, extend='max', alpha=0.9)
ax.scatter(trajectory_projection[0,32:-1], trajectory_projection[1,32:-1], c=['C1']*32, label='Component')
ax.scatter(trajectory_projection[0,-1:], trajectory_projection[1,-1:], c=['r'], marker='*', s=200.0, label='Mean')
ax.set_title('(b) Train loss')
ax.legend()
cbar = fig.colorbar(cs, ax=ax)
ax = axes[0][0]
cs = ax.contourf(xs, ys, train_err, cmap=cm.Spectral_r, 
                 levels = 20, extend='max', alpha=0.9)
ax.scatter(trajectory_projection[0,32:-1], trajectory_projection[1,32:-1], c=['C1']*32, label='Component')
ax.scatter(trajectory_projection[0,-1:], trajectory_projection[1,-1:], c=['r'], marker='*', s=200.0, label='Mean')
ax.set_title('(a) Train error (%)', fontsize=12)
ax.set_ylabel('Third component', fontsize=12)
cbar = fig.colorbar(cs, ax=ax)
ax = axes[1][0]
cs = ax.contourf(xs, ys, test_err, cmap=cm.Spectral_r, 
                 levels = 20, extend='max', alpha=0.9)
ax.scatter(trajectory_projection[0,32:-1], trajectory_projection[1,32:-1], c=['C1']*32, label='Component')
ax.scatter(trajectory_projection[0,-1:], trajectory_projection[1,-1:], c=['r'], marker='*', s=200.0, label='Mean')
ax.set_title('(c) Test error (%)', fontsize=12)
ax.set_ylabel('Third component', fontsize=12)
ax.set_xlabel('Second component', fontsize=12)
cbar = fig.colorbar(cs, ax=ax)
ax = axes[1][1]
cs = ax.contourf(xs, ys, test_loss, cmap=cm.Spectral_r, 
                 levels = 20, extend='max', alpha=0.9)
ax.scatter(trajectory_projection[0,32:-1], trajectory_projection[1,32:-1], c=['C1']*32, label='Component')
ax.scatter(trajectory_projection[0,-1:], trajectory_projection[1,-1:], c=['r'], marker='*', s=200.0, label='Mean')
ax.set_title('(d) Test loss', fontsize=12)
cbar = fig.colorbar(cs, ax=ax)
ax.set_xlabel("Second component", fontsize=12)
plt.tight_layout()
fig.suptitle("iBNN VGG-16 with 32 components on CIFAR-100", fontsize=15, y=1.)
plt.savefig('vgg_pca.jpg', bbox_inches='tight')
plt.show()
```
